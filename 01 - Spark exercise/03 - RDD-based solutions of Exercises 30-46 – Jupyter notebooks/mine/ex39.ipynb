```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
inputPath  = "../../data/Ex39/data/sensors.txt"
outputPath = "../../res/out_Ex39/" 
thresh = 50
```

```python
# Read the content of the input file
readingsRDD = sc.textFile(inputPath)
```

```python
graterRDD = readingsRDD.filter(lambda item : (float)(item.split(",")[2]) > 50)
```

```python
pairIDDATE = graterRDD.map(lambda item : (item.split(",")[0] , item.split(",")[1]))
```

```python
gouped = pairIDDATE.groupByKey()
gouped.collect()
```

```python
gouped.mapValues(lambda item : list(item)).collect()
```
