```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
inputPath  = "../../data/Ex41/data/sensors.txt"
outputPath = "../../res/out_Ex41/" 
thresh = 50
```

```python
# Read the content of the input file
readingsRDD = sc.textFile(inputPath)
```

```python
filteredRDD = readingsRDD.filter(lambda line : (float)(line.split(',')[2]) > 50)
```

```python

pairCount = filteredRDD.map(lambda line : (line.split(',')[0],1))
```

```python
Sumpaircoun = pairCount.reduceByKey(lambda value1,value2 : value1+value2)
```

```python
sorted = Sumpaircoun.sortBy(lambda item : item[1],False)
```

```python
sorted.collect()
```
