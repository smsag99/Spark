```python

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
inputPath  = "../../data/Ex37/data/sensors.txt"
outputPath = "../../res/out_Ex37/" 
```

```python
# Read the content of the input file
readingsRDD = sc.textFile(inputPath)
```

```python
pairIDPM = readingsRDD.map(lambda line : ( line.split(',')[0], (float) (line.split(',')[2])) )
```

```python
maxeachsensor = pairIDPM.reduceByKey(lambda value1,value2 : max(value1,value2) )
```

```python
maxeachsensor.collect()
```
