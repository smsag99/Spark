```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
inputPathQuestions  = "../../data/Ex42/data/questions.txt" # argv[1]
inputPathAnswers = "../../data/Ex42/data/answers.txt" # argv[2]
outputPath = "../res/out_Ex30/" 
```

```python
# Read the content of the question file
questionsRDD = sc.textFile(inputPathQuestions)
answersRDD = sc.textFile(inputPathAnswers)
```

```python
answersRDD.collect()
```

```python
pairIDQuestion = questionsRDD.map(lambda line: (line.split(',')[0],line.split(',')[2]))
pairIDAnswer= answersRDD.map(lambda line: (line.split(',')[1],line.split(',')[3]))

```

```python
pairIDQuestion.collect()
```

```python
joined = pairIDQuestion.join(pairIDAnswer)
```

```python
joined.collect()
```

```python
joined.groupByKey().map(lambda item : (item[0],list(item[1]))).collect()
```
