```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
inputPath  = "../../data/Ex38/data/sensors.txt"
outputPath = "../../res/out_Ex37/" 
thresh = 50
```

```python
# Read the content of the input file
readingsRDD = sc.textFile(inputPath)
```

```python
grater = readingsRDD.filter(lambda line : (float)(line.split(',')[2]) > thresh)
grater
```

```python
pairIDCount = grater.map(lambda item : (item.split(',')[0], 1))
pairIDCount.take(10)
```

```python
sunPairs = pairIDCount.reduceByKey(lambda value1,value2 : value1 + value2)
```

```python
sunPairs.filter(lambda item : item[1] >= 2).collect()
```
